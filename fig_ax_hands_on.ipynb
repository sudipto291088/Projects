```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[5,6,7])
```

```python
fig,ax = plt.subplots()


ax.plot([1,2,3],[5,6,7])
```

```python
fig,ax = plt.subplots(2,1)


ax[0].plot([1,2,3],[5,6,7])
ax[1].plot([2,2,5],[5,8,7])

plt.suptitle("Hands-on practice")
```

```python
fig,ax = plt.subplots(3,1,figsize = (10,10))


x = np.linspace(0,10,100)
ax[0].plot(x,np.sin(x),color = 'blue')
ax[1].plot(x,np.cos(x),color = 'green')
ax[2].plot(x,np.tan(x),color = 'red')


plt.tight_layout(pad=4)
```

```python
fig,ax = plt.subplots(2,2,figsize = (16,10))


x = np.linspace(0,10,100)
y = np.tan(x)
noise = np.random.randn(100)
data = np.random.randn(200)
categories = ['A','B','C','D']
values = [5,7,3,8]



##-----Top Left-----

ax[0,0].plot(x,y,color = 'blue')



##-----Top Right-----
ax[0,1].scatter(x,y + noise * 0.2,color = 'red', alpha = 0.9)



##-----Bottom Left-----
ax[1,0].bar(categories,values,color = 'orange')




##-----Bottom Right-----
ax[1,1].hist(data,bins=15,color='green')


fig.suptitle("2x2 Grid of Different Plot Types", fontsize=14, fontweight='bold')
plt.tight_layout(pad=2.0)
```
